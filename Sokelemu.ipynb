```python
import numpy as np
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as web
import yfinance as yfin
yfin.pdr_override()
plt.style.use('fivethirtyeight')
```

```python
#start = dt.datetime(2019,1,1)
end = dt.datetime.now()
start = end - dt.timedelta(days=1825)
```

```python
dataframe = pd.DataFrame()

#22 db
assets = ['INTC','BRK-B','BMRN','XOM','KO','VZ','T','SBUX','CLNE','FDX','AAL','BFOCX','UCTT','CAT',
          'NLY','PG','CSCO','NEE','FE','GLD','SPY','BAC']

for stock in assets:
    dataframe[stock] = web.get_data_yahoo(stock, start=start, end=end)["Close"]
```

```python
returns = np.log(dataframe/dataframe.shift(1))

cov_matrix = returns.cov()*252
corr_matrix = returns.corr()

plt.figure(figsize=(20,9))
plt.plot(returns, lw=2)
plt.xlabel('DATE', fontweight='bold')
plt.ylabel('RETURNS', fontweight='bold')
plt.legend(returns, fontsize=15, loc="upper left")
plt.show()
#returns.head()
```

```python
weights = np.random.random(len(dataframe.columns))
weights /= weights.sum()

portfolios = 10000

port_weights = np.zeros((portfolios, len(dataframe.columns)))
port_ret = np.zeros(portfolios)
port_risk = np.zeros(portfolios)
port_sharpe = np.zeros(portfolios)

for i in range(portfolios):
    weights = np.random.random(len(dataframe.columns))
    weights /= weights.sum()
    port_weights[i] = weights
    port_ret[i] = np.sum(returns.mean() * weights * 252)
    port_risk[i] = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    port_sharpe[i] = (port_ret[i] / port_risk[i])
    
plt.figure(figsize=(15,10))
plt.scatter(port_risk, port_ret, c=port_sharpe)

plt.scatter(port_risk[port_sharpe.argmax()],port_ret[port_sharpe.argmax()], c='red', s=125)
plt.scatter(port_risk[port_risk.argmin()],port_ret[port_risk.argmin()],c='black',s=125)
plt.title('EFFICIENT FRONTIER',fontweight='bold')
plt.xlabel('RISK (%)',fontweight='bold')
plt.ylabel('RETURN (%)',fontweight='bold')
plt.show()

print('Return: {:.2f}'.format(port_ret[port_sharpe.argmax()]*100),'%')
print('Risk: {:.2f}'.format(port_risk[port_sharpe.argmax()]*100),'%')
print('Min Return at Min Risk: {:.2f}'.format(port_ret[port_risk.argmin()]*100),'%')
print('Min Risk: {:.2f}'.format(port_risk[port_risk.argmin()]*100),'%')
print('Sharpe: {:.2f}'.format(port_sharpe.max()))
print('Weights: ',np.round(port_weights[port_sharpe.argmax()],4))
print('-------------------------------------------------------------------------------')
print(dataframe.columns[0]+' :',np.round(port_weights[port_sharpe.argmax(),0]*100,3),'%')
print(dataframe.columns[1]+':',np.round(port_weights[port_sharpe.argmax(),1]*100,3),'%')
print(dataframe.columns[2]+' :',np.round(port_weights[port_sharpe.argmax(),2]*100,3),'%')
print(dataframe.columns[3]+'  :',np.round(port_weights[port_sharpe.argmax(),3]*100,3),'%')
print(dataframe.columns[4]+'   :',np.round(port_weights[port_sharpe.argmax(),4]*100,3),'%')
print(dataframe.columns[5]+'   :',np.round(port_weights[port_sharpe.argmax(),5]*100,3),'%')
print(dataframe.columns[6]+'    :',np.round(port_weights[port_sharpe.argmax(),6]*100,3),'%')
print(dataframe.columns[7]+' :',np.round(port_weights[port_sharpe.argmax(),7]*100,3),'%')
print(dataframe.columns[8]+' :',np.round(port_weights[port_sharpe.argmax(),8]*100,3),'%')
print(dataframe.columns[9]+'  :',np.round(port_weights[port_sharpe.argmax(),9]*100,3),'%')
print(dataframe.columns[10]+'  :',np.round(port_weights[port_sharpe.argmax(),10]*100,3),'%')
print(dataframe.columns[11]+':',np.round(port_weights[port_sharpe.argmax(),11]*100,3),'%')
print(dataframe.columns[12]+' :',np.round(port_weights[port_sharpe.argmax(),12]*100,3),'%')
print(dataframe.columns[13]+'  :',np.round(port_weights[port_sharpe.argmax(),13]*100,3),'%')
print(dataframe.columns[14]+'  :',np.round(port_weights[port_sharpe.argmax(),14]*100,3),'%')
print(dataframe.columns[15]+'   :',np.round(port_weights[port_sharpe.argmax(),15]*100,3),'%')
print(dataframe.columns[16]+' :',np.round(port_weights[port_sharpe.argmax(),16]*100,3),'%')
print(dataframe.columns[17]+'  :',np.round(port_weights[port_sharpe.argmax(),17]*100,3),'%')
print(dataframe.columns[18]+'   :',np.round(port_weights[port_sharpe.argmax(),18]*100,3),'%')
print(dataframe.columns[19]+'  :',np.round(port_weights[port_sharpe.argmax(),19]*100,3),'%')
print(dataframe.columns[20]+'  :',np.round(port_weights[port_sharpe.argmax(),20]*100,3),'%')
print(dataframe.columns[21]+'  :',np.round(port_weights[port_sharpe.argmax(),21]*100,3),'%')
```

```python
#BETA
#Yahoo Finance-rol (kozeli ertekek)


rf = 1.48
rf_beta = 0

intc_beta = 0.51
brkb_beta = 0.91
bmrn_beta = 0.52
xom_beta = 1.34
ko_beta = 0.66
vz_beta = 0.41
t_beta = 0.66
sbux_beta = 0.83
clne_beta = 1.86
fdx_beta = 1.3
aal_beta = 1.68
bfocx_beta = 1.18
uctt_beta = 2.03
cat_beta = 0.91
nly_beta = 1.25
pg_beta = 0.41
csco_beta = 0.9
nee_beta = 0.26
fe_beta = 0.31
gld_beta = 0.09
spy_beta = 1
bac_beta = 1.54

spy_ret = np.sum(returns.loc[:, 'SPY'].mean()*252)*100

portfolio_beta = ((port_risk[port_sharpe.argmax()]**2))/cov_matrix.loc["SPY","SPY"]

betas = [rf_beta, intc_beta, brkb_beta, bmrn_beta, xom_beta, ko_beta, vz_beta, t_beta, sbux_beta, clne_beta,
        fdx_beta, aal_beta, bfocx_beta, uctt_beta, cat_beta, nly_beta, pg_beta, csco_beta, nee_beta,
        fe_beta, gld_beta, spy_beta, bac_beta]
```

```python
portfolio_beta = ((port_risk[port_sharpe.argmax()]**2))/cov_matrix.loc["SPY","SPY"]
#print(portfolio_beta)
#print(capm_portfolio)
#print(capm_spy)

capm_portfolio = rf + portfolio_beta * (spy_ret - rf)
print("Market return: ",round(capm_spy, 3),"%")
print("Portfolio return:", round(capm_portfolio,3),"%")
```

```python
#CAPM
#capm_pfe = rf + pfe_beta * (spy_ret - rf)
capm_rf = rf
capm_intc = rf + intc_beta * (spy_ret - rf)
capm_brkb = rf + brkb_beta * (spy_ret - rf)
capm_bmrn = rf + bmrn_beta * (spy_ret - rf)
capm_xom = rf + xom_beta * (spy_ret - rf)
capm_ko = rf + ko_beta * (spy_ret - rf)
capm_vz = rf + vz_beta * (spy_ret - rf)
capm_t = rf + t_beta * (spy_ret - rf)
capm_sbux = rf + sbux_beta * (spy_ret - rf)
capm_clne = rf + clne_beta * (spy_ret - rf)
capm_fdx = rf + fdx_beta * (spy_ret - rf)
capm_aal = rf + aal_beta * (spy_ret - rf)
capm_bfocx = rf + bfocx_beta * (spy_ret - rf)
capm_uctt = rf + uctt_beta * (spy_ret - rf)
capm_cat = rf + cat_beta * (spy_ret - rf)
capm_nly = rf + nly_beta * (spy_ret - rf)
capm_pg = rf + pg_beta * (spy_ret - rf)
capm_csco = rf + csco_beta * (spy_ret - rf)
capm_nee = rf + nee_beta * (spy_ret - rf)
capm_fe = rf + fe_beta * (spy_ret - rf)
capm_gld = rf + gld_beta * (spy_ret - rf)
capm_spy = rf + spy_beta * (spy_ret - rf)
capm_bac = rf + bac_beta * (spy_ret - rf)

capm_portfolio = rf + portfolio_beta * (spy_ret - rf)

capms = [capm_rf, capm_intc, capm_brkb, capm_bmrn, capm_xom, capm_ko, capm_vz, capm_t, capm_sbux, capm_clne, capm_fdx,
        capm_aal, capm_bfocx, capm_uctt, capm_cat, capm_nly, capm_pg, capm_csco, capm_nee, capm_fe, capm_gld,
        capm_spy, capm_bac]
```

```python
plt.figure(figsize=(20,9))

plt.scatter(x=sorted(betas), y=capms, label="CAPM stock returns", c="orange", lw=3)
plt.scatter(x=spy_beta, y=capm_spy, label="Market", c="blue", lw=6)
plt.scatter(x=portfolio_beta, y=capm_portfolio, label="Portfolio", c="red", lw=6)
plt.scatter(x=betas[0], y=capms[0], label="Risk Free Rate", c="black", lw=3)
plt.title("Capital Asset Pricing Model", fontweight="bold")
plt.xlabel("BETA", fontweight="bold")
plt.ylabel("RETURN(%)", fontweight="bold")

z = np.polyfit(betas, capms, 1)
p = np.poly1d(z)
plt.plot(betas, p(betas), "--" ,lw=2, label="Security Market Line")
plt.legend()
plt.show()
```
