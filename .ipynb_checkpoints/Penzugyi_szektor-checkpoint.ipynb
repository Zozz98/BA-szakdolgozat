```python
import numpy as np
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt.discrete_allocation import DiscreteAllocation
from pypfopt.discrete_allocation import get_latest_prices
plt.style.use('fivethirtyeight')
```

```python
start = dt.datetime(2010,1,1)
end = dt.datetime.now()
```

```python
#Bank of America - BAC
#Morgan Stanley - MS
#Wells Fargo - WFC
#Financial Select Sector - XLF
#Citigroup - C

it_szektor = ['BAC', 'MS', 'WFC', 'XLF', 'C']

dataframe = pd.DataFrame()

for stock in it_szektor:
    dataframe[stock] = web.DataReader(stock, data_source='yahoo', start=start, end=end)['Close']
```

```python
plt.figure(figsize=(15,10))

for t in dataframe.columns.values:
    plt.plot(dataframe[t], label = t, lw=2)
    plt.title('PORTFOLIO RESZVENYEI')


plt.xlabel('DATUM')
plt.ylabel('ARFOLYAM($)')
plt.legend(fontsize=15)
plt.show()
```

```python
returns = dataframe.pct_change()
returns.plot(alpha=0.5, figsize=(15,10), lw=2)
plt.title('RESZVENYEK SZORASA')
plt.xlabel('DATUM')
plt.ylabel('SZORAS')
plt.show()
```

```python
# ~250 nap van egy evben
kov_matrix = returns.cov() * 250 
kov_matrix
```

```python
korr_matrix = returns.corr()
korr_matrix
```

```python
#Bank of America-ba:
bac_ret = np.sum(returns.loc[:, 'BAC'].mean()*250)*100
bac_vol = np.sqrt(kov_matrix.iloc[0,0])*100
print('A BAC hozama: {:.2f}'.format(bac_ret)+'%')
print('A BAC szorasa: {:.2f}'.format(bac_vol)+'%')

print('-----------------------')

#Morgan Stanley-be:
ms_ret = np.sum(returns.loc[:, 'MS'].mean()*250)*100
ms_vol = np.sqrt(kov_matrix.iloc[1,1])*100
print('Az MS hozama: {:.2f}'.format(ms_ret)+'%')
print('Az MS szorasa: {:.2f}'.format(ms_vol)+'%')

print('-----------------------')

#Wells Fargo-ba:
wfc_ret = np.sum(returns.loc[:, 'WFC'].mean()*250)*100
wfc_vol = np.sqrt(kov_matrix.iloc[2,2])*100
print('A WFC hozama: {:.2f}'.format(wfc_ret)+'%')
print('A WFC szorasa: {:.2f}'.format(wfc_vol)+'%')

print('-----------------------')

#Financial Select Sector-ba:
xlf_ret = np.sum(returns.loc[:, 'XLF'].mean()*250)*100
xlf_vol = np.sqrt(kov_matrix.iloc[3,3])*100
print('Az XLF hozama: {:.2f}'.format(xlf_ret)+'%')
print('Az XLF szorasa: {:.2f}'.format(xlf_vol)+'%')

print('-----------------------')

#Citigroup-ba:
c_ret = np.sum(returns.loc[:, 'C'].mean()*250)*100
c_vol = np.sqrt(kov_matrix.iloc[4,4])*100
print('A C hozama: {:.2f}'.format(c_ret)+'%')
print('A C szorasa: {:.2f}'.format(c_vol)+'%')
```
