```python
import numpy as np
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt.discrete_allocation import DiscreteAllocation
from pypfopt.discrete_allocation import get_latest_prices
plt.style.use('fivethirtyeight')
```

```python
start = dt.datetime(2010,1,1)
end = dt.datetime.now()
```

```python
#Canadian Solar - CSIQ
#iClean Global Clean Energy - ICLN
#NextEra Energy - NEE
#First Solar - FSLR
#SunPower Corp - SPWR

it_szektor = ['CSIQ', 'ICLN', 'NEE', 'FSLR', 'SPWR']

dataframe = pd.DataFrame()

for stock in it_szektor:
    dataframe[stock] = web.DataReader(stock, data_source='yahoo', start=start, end=end)['Close']
```

```python
plt.figure(figsize=(15,10))

for t in dataframe.columns.values:
    plt.plot(dataframe[t], label = t, lw=2)
    plt.title('PORTFOLIO RESZVENYEI')


plt.xlabel('DATUM')
plt.ylabel('ARFOLYAM($)')
plt.legend(fontsize=15)
plt.show()
```

```python
returns = dataframe.pct_change()
returns.plot(alpha=0.5, figsize=(15,10), lw=2)
plt.title('RESZVENYEK SZORASA')
plt.xlabel('DATUM')
plt.ylabel('SZORAS')
plt.show()
```

```python
# ~250 nap van egy evben
kov_matrix = returns.cov() * 250 
kov_matrix
```

```python
korr_matrix = returns.corr()
korr_matrix
```

```python
#Canadian Solar-ba:
csiq_ret = np.sum(returns.loc[:, 'CSIQ'].mean()*250)*100
csiq_vol = np.sqrt(kov_matrix.iloc[0,0])*100
print('A CSIQ hozama: {:.2f}'.format(csiq_ret)+'%')
print('A CSIQ szorasa: {:.2f}'.format(csiq_vol)+'%')

print('-----------------------')

#iClean Global Clean Energy-be:
icln_ret = np.sum(returns.loc[:, 'ICLN'].mean()*250)*100
icln_vol = np.sqrt(kov_matrix.iloc[1,1])*100
print('Az ICLN hozama: {:.2f}'.format(icln_ret)+'%')
print('Az ICLN szorasa: {:.2f}'.format(icln_vol)+'%')

print('-----------------------')

#NextEra-ba:
nee_ret = np.sum(returns.loc[:, 'NEE'].mean()*250)*100
nee_vol = np.sqrt(kov_matrix.iloc[2,2])*100
print('A NEE hozama: {:.2f}'.format(nee_ret)+'%')
print('A NEE szorasa: {:.2f}'.format(nee_vol)+'%')

print('-----------------------')

#First Solar-ba:
fslr_ret = np.sum(returns.loc[:, 'FSLR'].mean()*250)*100
fslr_vol = np.sqrt(kov_matrix.iloc[3,3])*100
print('Az FSLR hozama: {:.2f}'.format(fslr_ret)+'%')
print('Az FSLR szorasa: {:.2f}'.format(fslr_vol)+'%')

print('-----------------------')

#SunPower-be:
spwr_ret = np.sum(returns.loc[:, 'SPWR'].mean()*250)*100
spwr_vol = np.sqrt(kov_matrix.iloc[4,4])*100
print('Az SPWR hozama: {:.2f}'.format(spwr_ret)+'%')
print('Az SPWR szorasa: {:.2f}'.format(spwr_vol)+'%')
```
